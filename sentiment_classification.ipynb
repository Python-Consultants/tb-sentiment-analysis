```python
#读取情感词典的褒义词和贬义词
positive_comment_raw = pd.read_table('sentiment/positive_comment.txt',sep='\n', encoding='gbk')
positive_comment_raw.columns = ['word']
negative_comment_raw = pd.read_table('sentiment/negative_comment.txt',sep='\n', encoding='gbk')
negative_comment_raw.columns = ['word']
level_raw =  pd.read_table('sentiment/level.txt', sep='\n', encoding='gbk')
level_raw.columns = ['word']
level_raw.drop(0, inplace= True)
```

```python
positive_list = list(positive_comment_raw['word'])
negative_list = list(negative_comment_raw['word'])
for i in range(len(positive_list)):
    positive_list[i] = positive_list[i].replace(' ','')
for j in range(len(negative_list)):
    negative_list[j] = negative_list[j].replace(' ','')
```

```python
import xlrd
#读取停用词表
cutword_raw = xlrd.open_workbook('stopword20181231.xlsx')
table = cutword_raw.sheet_by_index(0) #通过索引顺序获取,0表示第一张表
cutword_list = [table.cell(i,0).value for i in range(0, 3366)] #共3366个
# 停用词表
CUTWORD_NOREP = list(set(cutword_list)) #对停用词做去重
for i in CUTWORD_NOREP:
    #词典中的词不去除
    if i in list(level_raw['word']):
        CUTWORD_NOREP.remove(i)
    elif i in positive_list:
        CUTWORD_NOREP.remove(i)
    elif i in negative_list:
        CUTWORD_NOREP.remove(i)
```

## 打分

```python
import jieba
import re

def cutword(old_description): #将分词后的某一个app对应的描述的词语列表去除无效词并去除单字
    new_description = []
    for i in range(len(old_description)):
        #可选择去除单字，仅保留词语，若要做这个选择，可以加上#后面的代码即可
        if (old_description[i] not in CUTWORD_NOREP): #and len(old_description[i]) > 1): 
            new_description.append(old_description[i])
    return (new_description)

def rawstr_process(raw_str): #对文本处理输出成词列表
    temp_str = re.sub('[^\w\u4e00-\u9fff]+', '',raw_str) #去除文本中的标点符号
    temp_str1 = re.sub("[A-Za-z0-9]", "", temp_str) #去除文本中的英文字母和数字
    #cut_all参数为jieba中可选的全分词模式，此模式可能会出现词语重合
    raw_jieba_list = list(jieba.cut(temp_str1,cut_all=False, HMM= False)) 
    return(cutword(raw_jieba_list))
```

```python
#计算得分
def score_calculation(processed_list):
    strong_positive_count = 0 
    positive_count = 0 
    strong_negative_count = 0
    negative_count = 0
    for i in processed_list:
        if i in positive_list:
            positive_count +=1
        elif i in negative_list:
            negative_count +=1
    if ((positive_count + negative_count) ==0):
        final_calculation = 'cannotDecide'
    else: final_calculation = (positive_count - negative_count)/(positive_count + negative_count)
    return(final_calculation)
```

```python
#sample
print(score_calculation(rawstr_process('宝贝收到了，质量非常好，与卖家描述的完全一致，品质保证！非常满意，真的很喜欢，完全超出期望值')))
print(score_calculation(rawstr_process('宝贝收到了，质量非常差，包装都碎了')))
print(score_calculation(rawstr_process('帮朋友买的')))
```

```python
#对整个数据集进行生产
total_text_list  = []
score_list = []
for i in range(len(raw_data_select)):
    processed_str = rawstr_process(comment_raw[i])
    total_text_list.append(processed_str)
    score_list.append(score_calculation(processed_str))

raw_data_select['score']=score_list
```
